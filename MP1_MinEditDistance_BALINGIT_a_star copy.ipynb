```python
def minimum_edit_distance(source, target):
    n = len(source)
    m = len(target)

    dp = [[0 for cell in range(m + 1)] for cell in range(n + 1)]


    # 1. initialize first row and column
    # `D(i, 0)`
    for i in range(1, n + 1):
        dp[i][0] = i                                # i deletions cost = 1
    # `D(0, j)`
    for j in range(1, m + 1):
        dp[0][j] = j                                # j insertions cost = 1


    # 2. fill up the DP table with a recurrence
    for i in range(1, n + 1):
        for j in range(1, m + 1):
            # dp[i][j] = dp[i-1][j] + 1     # delete
            # dp[i][j] = dp[i][j-1] + 1     # insert

            # if source[j - 1] != target[i - 1]:
            # elif source[j - 1] == target[i - 1]:
            # else:

            if source[i - 1] == target[j - 1]:
                dp[i][j] = dp[i - 1][j - 1]         # match cost = 0
            else:
                dp[i][j] = min(
                    dp[i - 1][j] + 1,      # Deletion
                    dp[i][j - 1] + 1,      # Insertion
                    dp[i - 1][j - 1] + 2   # Substitution (cost 2: Delete + Insert)
                )

    # import pandas as pd
    # dp_df = pd.DataFrame(dp)
    # dp_df


    # 3. backtrace to build alignment
    src_aligned = []
    tgt_aligned = []
    ops = []  

    i, j = n, m
    while i > 0 or j > 0:           # ai generated: debugged while loop
        current_cost = dp[i][j]

        # 3.1. check first if can move diagonally up left
            # if characters are same AND cost didn't change ==> MATCH
            # else if characters are NOT same AND cost increased by 2 ==> MISMATCH
        if i > 0 and j > 0:
            is_match = (source[i - 1] == target[j - 1])
            diagonal_cost = dp[i - 1][j - 1]
            
            # if cost + 0 ==> MATCH
            if is_match and current_cost == diagonal_cost:
                src_aligned.append(source[i - 1])
                tgt_aligned.append(target[j - 1])
                ops.append(' ') # append space character for match
                i -= 1
                j -= 1
                continue
            
            # if cost + 2 ==> MISMATCH
            elif not is_match and current_cost == diagonal_cost + 2:
                src_aligned.append(source[i - 1])
                tgt_aligned.append(target[j - 1])
                ops.append('M') # append M for mismatch
                i -= 1
                j -= 1
                continue

        # 3.2. if cannot move up left, check horizontal left j-1
        if j > 0 and current_cost == dp[i][j - 1] + 1:
            src_aligned.append('-')
            tgt_aligned.append(target[j - 1])
            ops.append('I') # append I for insert
            j -= 1
            
        # 3.3. if cannot move up left nor left, move up i-1
        elif i > 0 and current_cost == dp[i - 1][j] + 1:
            src_aligned.append(source[i - 1])
            tgt_aligned.append('-')
            ops.append('D') # append D for delete
            i -= 1

    src_aligned.reverse()
    tgt_aligned.reverse()
    ops.reverse()

    # 4. return back the dp array, the aligned source-target arrays, and the operations on the strings
    return dp[n][m], ''.join(src_aligned), ''.join(ops), ''.join(tgt_aligned)

#   # your code here
#   pass

# # Note: you're not forced to use this template, so you can implement multiple
# #       functions or rename functions as you see fit.
```

# A* Algorithm
## time complexity: `O(nm)` but `Î©(k)` for similar strings
instead of the dp's constant `O(nm)`


This solution treats the problem as a graph traversal, to reach the goal faster without visiting every possible state. 

Instead of using a 2D matrix, I've used a heap/priority queue instead.

```python
import heapq

def minimum_edit_distance(s1, s2):
    m, n = len(s1), len(s2)
    # heap stores: (f_score, g_score, i, j)
    # i and j are current indices in s1 and s2
    start_h = abs(m - n)
    open_set = [(start_h, 0, 0, 0)]
    
    # To store the path: (i, j) -> (prev_i, prev_j, operation)
    came_from = {}
    # To track the best g_score for each state
    g_score = {(0, 0): 0}

    while open_set:
        f, cost, i, j = heapq.heappop(open_set)

        # Goal check
        if i == m and j == n:
            return reconstruct_path(came_from, i, j), cost

        # Define possible moves: (di, dj, cost, op_name)
        moves = []
        if i < m: moves.append((1, 0, 1, "Delete"))       # Delete from s1
        if j < n: moves.append((0, 1, 1, "Insert"))       # Insert from s2
        if i < m and j < n:
            sub_cost = 0 if s1[i] == s2[j] else 2
            moves.append((1, 1, sub_cost, "Substitute" if sub_cost == 2 else "Keep"))

        for di, dj, move_cost, op in moves:
            ni, nj = i + di, j + dj
            new_g = cost + move_cost
            
            if (ni, nj) not in g_score or new_g < g_score[(ni, nj)]:
                g_score[(ni, nj)] = new_g
                # Heuristic: remaining length difference
                h = abs((m - ni) - (n - nj))
                f = new_g + h
                heapq.heappush(open_set, (f, new_g, ni, nj))
                came_from[(ni, nj)] = (i, j, op)

def reconstruct_path(came_from, i, j):
    path = []
    curr = (i, j)
    while curr in came_from:
        prev_i, prev_j, op = came_from[curr]
        path.append(f"{op} ({prev_i},{prev_j}) -> ({curr[0]},{curr[1]})")
        curr = (prev_i, prev_j)
    return path[::-1]
```
